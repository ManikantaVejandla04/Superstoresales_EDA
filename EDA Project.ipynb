## Phase-2 : Data Cleaning, Manipulation & Visualization

1.1- Load dataset and check missing values

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("SuperstoreData.csv")
df
```

```python
# Check for missing values

print(df.isnull().sum())
print("\nPercentage of missing values:")
print((df.isnull().sum() / len(df)) * 100)
```

- This code identifies which columns have null entries and shows the percentage of missing data in each column.
- Missing values can distort calculations and lead to incorrect business insights.
- Early detection allows you to choose the right imputation strategy (delete, fill with median, mode, or forward fill).

1.2- Visualize missing data pattern

```python
# Check missing values visually

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
missing_data = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({
    'Column': missing_data.index,
    'Missing_Count': missing_data.values,
    'Missing_Percent': missing_percent.values
})
print(missing_df[missing_df['Missing_Count'] > 0])
```

- Shows which columns have missing data and how many rows are affected.
- Helps prioritize which columns need immediate attention.

1.3- Handle missing values – Strategy for numeric columns

```python
# For numeric columns: Fill with MEDIAN (less affected by outliers)

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)
        print(f"Filled {col} with median value: {df[col].median()}")
```

- Median is preferred over mean for numeric data because it is robust to outliers.
- For Sales and Profit columns: median preserves realistic transaction values without extreme distortion.
- For Quantity and Discount: median imputation maintains the natural distribution of order patterns.

1.4- Handle missing values – Strategy for categorical columns

```python
# For categorical columns: Fill with MODE (most frequent value)

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)
        print(f"Filled {col} with mode: {df[col].mode()[0]}")
```

- Mode is the most frequently occurring category, maintaining the real distribution of data.
- For Region: filling with mode ensures geographic representation stays consistent with actual sales patterns.
- For Segment: mode imputation preserves the customer segment distribution without introducing artificial categories.

1.5-  Verify all missing values are handled

```python
# Confirm no missing values remain

print("Missing values after handling:")
print(df.isnull().sum().sum())  # Should be 0
print("\nDataset shape after cleaning:", df.shape)
```

- This confirms that all missing values have been successfully imputed.
- The dataset is now ready for analysis without incomplete records affecting results.

**2.** Remove Duplicates

2.1- Check for duplicate rows

```python
# Check for complete duplicates (all columns identical)

duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows: {duplicate_count}")

# Show a few duplicate rows if they exist

if duplicate_count > 0:
    print("\nSample duplicates:")
    print(df[df.duplicated(keep=False)].head(10))
```

- Detects if the entire row is repeated with identical values in all columns.
- Duplicates can occur from system errors, data import failures, or accidental entries.
- Finding duplicates before analysis prevents inflated statistics and misleading insights.

2.2- Check for duplicates based on key columns

```python
# Check duplicates based on Order_ID (if orders should be unique)

if 'Order_ID' in df.columns:
    order_duplicates = df[df.duplicated(subset=['Order_ID'], keep=False)]
    print(f"Duplicate Order IDs: {len(order_duplicates)}")
    if len(order_duplicates) > 0:
        print(order_duplicates.head())
```

- Some columns like Order_ID should be unique identifiers; duplicates here indicate data quality issues.
- Checking key columns separately helps identify logical duplicates (same order entered twice) vs. coincidental duplicates.

2.3- Remove duplicate rows

```python
# Remove complete duplicates (keep first occurrence)

df_cleaned = df.drop_duplicates(keep='first')

print(f"Rows removed: {len(df) - len(df_cleaned)}")
print(f"Dataset shape after removing duplicates: {df_cleaned.shape}")

# Update the original dataframe

df = df_cleaned
```

- keep='first' retains the first occurrence and removes subsequent duplicates.
- Duplicate removal prevents overcounting in sales/profit totals and customer metrics.
- After removal, each row represents a unique transaction or record.

2.4- Verify duplicates are removed

```python
# Confirm duplicates are gone

print("Remaining duplicates after cleaning:", df.duplicated().sum())
```

- This verification ensures the duplicate removal was successful.
- The dataset now has only unique records for accurate analysis.

**3.** Fix Inconsistencies in Categorical Data

3.1- Identify inconsistencies in categorical columns

```python
# Check unique values in categorical columns

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    print(f"\n{col}: {df[col].nunique()} unique values")
    print(df[col].value_counts())
```

- This shows all distinct values in text columns to spot inconsistencies.
- Variations like "Office Supplies", "office supplies", "OFFICE SUPPLIES" will appear as separate categories.
- Identifying these early prevents fragmentation in group-by analyses.

3.2- Fix case inconsistencies (uppercase/lowercase)

```python
# Standardize text to consistent case (lowercase)

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].str.lower()  # Convert to lowercase
        df[col] = df[col].str.strip()  # Remove leading/trailing spaces

print("Categorical columns standardized to lowercase and trimmed.")
```

- Converting to lowercase ensures "Yes", "yes", and "YES" are treated as one category.
- .str.strip() removes accidental spaces that prevent grouping.
- Standardization improves aggregation accuracy in pivot tables and group-by operations.

3.3: Check for leading/trailing whitespace

```python
# Double-check for any remaining whitespace issues

for col in categorical_cols:
    spaces = df[col].str.contains('^ | $', regex=True, na=False).sum()
    if spaces > 0:
        print(f"{col} still has {spaces} entries with leading/trailing spaces")
        df[col] = df[col].str.strip()
```

- Whitespace is invisible but prevents proper matching and grouping.
- Removing it ensures "Region " and "Region" are recognized as the same category.

3.4- Unify category names

```python
# Example: Unify Category variations

if 'Category' in df.columns:
    # Replace common variations
    df['Category'] = df['Category'].replace({
        'office supplies': 'office supplies',
        'technology': 'technology',
        'furniture': 'furniture'
    })

# Check final unique values

print("\nFinal unique values in categorical columns:")
for col in categorical_cols:
    print(f"{col}: {df[col].unique()}")
```

- Ensures all category names are consistent and standardized.
- Prevents analysis from treating the same product category as multiple groups.
- Clean categories improve the accuracy of segment and region-wise summaries.

**4.** Convert Data Types

4.1- Check current data types

```python
# Display current data types

print("Current data types:")
print(df.dtypes)
```

- Shows which columns are strings (object), numbers (int64, float64), or other types.
- Incorrect types waste memory and slow down calculations.

4.2- Convert date columns to datetime

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='mixed')

# Save converted file

df.to_csv("SuperstoreData.csv", index=False)
```

```python
df.dtypes
```

- Converting to datetime enables time-based analysis (e.g., sales trends over months, days between order and ship).
- Datetime format allows sorting, filtering by date ranges, and extracting year/month/day.
- Improves memory efficiency and enables advanced time series operations.

4.3- Convert numerical strings to numbers

```python
# Example: Ensure numeric columns are truly numeric

numeric_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to float; invalid entries become NaN
        print(f"Converted {col} to numeric")
```

- pd.to_numeric() converts strings like "100.50" to actual numbers for calculations.
- errors='coerce' converts problematic values to NaN so they can be handled appropriately.
- Numeric types enable mathematical operations like sum, mean, correlation.

4.4- Convert categorical columns to category type (optional, for memory efficiency)

```python
# Mark low-cardinality columns as 'category' for memory efficiency

categorical_cols_to_convert = ['Category', 'Sub-Category', 'Region', 'Segment', 'Country']

for col in categorical_cols_to_convert:
    if col in df.columns:
        df[col] = df[col].astype('category')
        print(f"Converted {col} to category type")

print("\nFinal data types:")
print(df.dtypes)
```

- Category type reduces memory usage for columns with many repeated values.
- Improves performance in groupby and pivot operations.
- Makes intent clear: this column represents discrete groups, not free text.

4.5- Verify conversion success

- Confirms all conversions were successful.
